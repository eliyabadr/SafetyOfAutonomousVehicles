```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
m = 1
g = 9.81
v0 = 10
T_h = 0.5
#assumed cg = ca = 1
```

```python
def F_r(v):
    f0, f1, f2 = 0.1, 0.01, 0.001
    return f0 + f1*v + f2*v**2
```

```python
def cbf(z,v):
    return z - T_h*v - (1/2)*((v0-v)**2)/g
```

```python
v_vals = np.linspace(0.00, 15, 200)   # ego speed (v)
z_vals = np.linspace(0.01, 20, 200)   # gap (z)
V, Z = np.meshgrid(v_vals, z_vals)
```

```python
cbf_values = cbf(Z,V)
print(cbf_values) #we can see negative and positive values
```

```python
plt.figure(figsize=(8, 6))
vmin, vmax = -2, 2
contourf = plt.contourf(V, Z, cbf_values, levels=np.linspace(vmin, vmax, 50), cmap='viridis', extend='both')
plt.colorbar(contourf, label='CBF B(x)')
contour = plt.contour(V, Z, cbf_values, levels=[0], colors='red', linewidths=2)

plt.xlabel('Velocity v (m/s)')
plt.ylabel('Gap z (m)')
plt.title(f"Safety Boundary from CBF (Th = {T_h}s, v0 = {v0})")
plt.grid(True)
plt.tight_layout()
plt.show()
```
